```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
volts = [3.112, 4.50, 5.28, 5.78, 6.14, 6.40]
amps = [0.006224, 0.0045, 0.00352, 0.00289, 0.002456, 0.002133]
```

```python
fig = plt.figure()

axes = fig.add_axes([1, 1, 1, 1])

axes.set_xlim([0, 0.01])
axes.set_ylim([0, 10])

axes.set_title("Figure 1: Box 2 Voltage vs Current")
axes.set_xlabel("Current (Amperes)")
axes.set_ylabel("Voltage (Volts)")
axes.scatter(amps, volts)

slope, y_int = np.polyfit(amps, volts, 1)

arr = np.linspace(0, 0.1, 1000)

print("The slope is %s." % (slope))
print("The y-intercept is %s." % (y_int))

yVals = (arr * slope) + y_int
axes.plot(arr, yVals)
```
