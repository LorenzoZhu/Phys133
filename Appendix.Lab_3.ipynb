```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
f = plt.figure(figsize=(7,7))
V = (3.11,4.50,5.28,5.78,6.14,6.40)
I = np.array([0.00622,0.00450,0.00352,0.00289,0.00246,0.00213])
plt.scatter(I,V)

def func(x,a,b):
    return a*x + b

R0 ,V0 = np.poly1d(np.polyfit(I, V, 1))
plt.plot(I,func(I,R0,V0))
plt.xlim(0.001,0.0075)

plt.title("Figure 3")
plt.xlabel("I")
plt.ylabel("V")
plt.savefig("DC box2.png")
print(-R0,V0)
```

### Box B

```python
f_B = np.array([10,100,300,500,700,900,1000,2000,3000,5000,6000,7000,8000,10000,20000,30000,50000,70000,80000])
C1_B = np.array([10.2,10.0,10.2,10.7,11.4,13.2,12.7,18.8,26.4,33.6,36.4,39.2,42.4,45.1,48.0,49.6,49.0,49.2,49.0])
C2_B = np.array([37.6,37.2,36.8,36.8,36.4,38.0,36.4,35.2,36.4,30.0,28.4,26,42.4,22.0,12.2,8.40,4.45,3.12,2.68])
phase_B = np.array([-177,-177,-166,-156,-150,-144,-142,-125,-115,-108,-105,-104,-103,-103,-95,-94,-90,-90.2,-90.1,-89])
Z_B = 200*(C1_B/C2_B)
```

```python
#plot angular frequency vs impedience
#from this plot we know it's a resistor and inductor in series
#thus the function would be Z^2 = R^2 + (iwL)^2

f = plt.figure(figsize=(7,7))
plt.scatter((2*np.pi*f_B), Z_B)

plt.title("Figure 4")
plt.xlabel("ω")
plt.ylabel("Z")
plt.savefig("B-w-Z.png")
```

```python
#the model functiuon is y = ax+b
#where a is L^2 and b is R^2, y=Z^2 and x=w^2
def funcB(x,a,b):
    return a*x + b

ZB_sqr = Z_B**2
wB_sqr = (2*np.pi*f_B)**2
```

```python
#plot w^2 vs Z^2
#fit a line
f = plt.figure(figsize=(7,7))
plt.scatter(wB_sqr, ZB_sqr)
LB ,b = np.poly1d(np.polyfit(wB_sqr, ZB_sqr, 1))
print("Inductor value",np.sqrt(LB))
plt.plot(wB_sqr,funcB(wB_sqr,LB,54.3**2))

plt.title("Figure 5")
plt.xlabel("ω^2")
plt.ylabel("Z^2")
plt.savefig("B-fit.png")
```

```python
f_F = np.array([50,100,300,500,700,900,1000,3000,5000,7000,9000,10000,20000,30000,40000,50000,70000,80000,90000])
C1_F = np.array([45.2,45.2,45.6,45.6,45.2,45.2,45.3,46.5,46.4,46.4,46.2,46.2,45.3,44.2,42.4,40.6,36.8,34.6,32.8])
C2_F = np.array([4.16,4.20,4.20,4.16,4.22,4.24,4.30,4.60,5.16,6.12,6.96,7.40,12.5,18.0,22.6,26.6,32.8,34.6,36.4])
phase_F = np.array([-179,-179,179,178,176,175,175,158,149,139,135,134,116,110,106,105,100,101,100,101])
Z_F = 200*(C1_F/C2_F)
```

```python
#resistor and capactor in parrell
f = plt.figure(figsize=(7,7))
wF = 2*np.pi*f_F
plt.scatter(wF, Z_F)

plt.title("Figure 6")
plt.xlabel("ω")
plt.ylabel("Z")
plt.savefig("F-w-Z.png")
```

```python
#the model functiuon is y = ax+b
#where a is C^2 and b is 1/R^2, y=Y^2 and x=w^2
def funcF(x,a,b):
    return a*x + b

YF_sqr = 1/Z_F**2
wF_sqr = wF**2
```

```python
#plot w^2 vs 1/Z^2
#fit a line
f = plt.figure(figsize=(7,7))
plt.scatter(wF_sqr,YF_sqr)
CF ,RF = np.poly1d(np.polyfit(wF_sqr, YF_sqr, 1))
print("capacitor value",np.sqrt(CF),"resistor value",1/np.sqrt(np.abs(RF)))
plt.plot(wF_sqr,funcF(wF_sqr,CF,1/2187**2))
plt.ylim(-0.000001,0.00004)

plt.title("Figure 7")
plt.xlabel("ω^2")
plt.ylabel("Y^2")
plt.savefig("F-fit.png")
```

```python
Zlow_E = 100000*np.array([2.76,2.58,2.38,2.17,2.05,1.66])/np.array([0.872,1.30,1.67,1.93,2.04,2.32])
Zhigh_E = 100000*np.array([0.58,1.12,2.20,4.80,14.6])/np.array([3.52,4.16,5.28,7.68,17.0])
f_lowE = np.array([300,500,700,900,1000,1500])
f_highE = np.array([60000,70000,80000,90000,100000])
w_lowE = f_lowE*2*np.pi
w_highE = f_highE*2*np.pi
```

```python
#the model functiuon is y = ax+b
#where a is 1/C^2 and b is R^2, y=Z^2 and x=1/w^2
def lowE(x,a,b):
    return a*x + b

#here a is L^2 and b is R^2, y=Z^2 and x=w^2
def highE(x,c,d):
    return c*x + d
```

```python
f = plt.figure(figsize=(7,7))
plt.scatter(1/w_lowE**2,Zlow_E**2)
CE ,RE = np.poly1d(np.polyfit(1/w_lowE**2,Zlow_E**2, 1))
print("capacitor and resistor value",1/np.sqrt(CE))
print(np.sqrt(RE))
plt.plot(1/w_lowE**2,lowE(1/w_lowE**2,CE,RE))
plt.xlim(0,3e-7)

plt.title("Figure 9")
plt.xlabel("ω^2")
plt.ylabel("Z^2")
plt.savefig("E-low-fit.png")
```

```python
f = plt.figure(figsize=(7,7))
plt.scatter(w_highE**2,Zhigh_E**2)
LE ,RE_L = np.poly1d(np.polyfit(w_highE**2,Zhigh_E**2, 1))
print("inductor and resistor value",np.sqrt(LE),np.sqrt(np.abs(RE_L)))
plt.plot(w_highE**2,highE(w_highE**2,LE,RE_L))

plt.title("Figure 10")
plt.xlabel("ω^2")
plt.ylabel("Z^2")
plt.savefig("E-high-fit.png")
```
